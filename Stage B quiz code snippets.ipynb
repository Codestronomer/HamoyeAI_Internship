```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')

df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
# Takes T2 from the dataframe
X = df[['T2']]

# Takes T6 from the dataframe
y = df[['T6']]
```

```python
from sklearn.model_selection import train_test_split

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y)
```

```python
# Checks shape of training data set
X_train_1.shape, y_train_1.shape
```

```python
# Checks shape of test data set
X_test_1.shape, y_test_1.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()

# Trains the model
model1.fit(X_train_1, y_train_1)
```

```python
# predicts with trained model
pred = model1.predict(X_test_1)
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(pred, y_test_1)
round(accuracy, 2)
```

```python
# Removes date and lights from the dataset
df2 = df.drop(['date', 'lights'], axis=1)
df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

# Normalize the dataset
normalized_df = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
normalized_df.head()
```

```python
featured_df = normalized_df.drop(['Appliances'], axis=1)

# Get the target variable from the normalized dataset
target = normalized_df[['Appliances']]
target.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(featured_df, target, train_size=0.7, random_state=42)
```

```python
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
pred_values = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# Mean absolute error
mae = mean_absolute_error(y_test, pred_values)
round(mae, 2)
```

```python
# Residual sum of squares
rss = np.sum(np.square(y_test - pred_values))
round(rss, 2)
```

```python
# Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, pred_values))
round(rmse, 3)
```

```python
# Coefficient of Determination

from sklearn.metrics import r2_score
c_o_d = r2_score(y_test, pred_values)
round(c_o_d, 2)
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)

# Lasso Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso()
lasso_reg.fit(X_train, y_train)

# Comparing the effect of regularisation
def df_weights(model, feat, col_name):
    # this function returns the weight of every feature
    weights = pd.Series(model.coef_[0], feat.columns)
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = df_weights(model, X_train, 'Linear_model_weight')
ridge_weights_df = df_weights(ridge_reg, X_train, 'Ridge_Weight')
lasso_weights_df = df_weights(lasso_reg, X_train, 'Lasso Weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```

```python
# Ridge RMSE
ridge_pred = ridge_reg.predict(X_test)
ridge_rmse = np.sqrt(mean_squared_error(y_test, pred_values))
round(ridge_rmse, 3)
```

```python
# Lasso Regression
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)


# Comparing the effect of regularisation
def df_weights(model, feat, col_name):
    # this function returns the weight of every feature
    weights = pd.Series(model.coef_[0], feat.columns)
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

lasso_weights_df = df_weights(lasso_reg, X_train, 'Lasso Weight')
lasso_weights_df
```

```python
# Lasso RMSE
lasso_pred = lasso_reg.predict(X_test)
lasso_rmse = np.sqrt(mean_squared_error(y_test, pred_values))
round(lasso_rmse, 3)
```
