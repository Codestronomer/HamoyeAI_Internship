```python
import numpy as np
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv")

df.head()
```

```python
df['stabf'].value_counts(normalize=True)
```

```python
stability = {'unstable': 0, 'stable': 1}
df['stabf'] = df['stabf'].replace(stability)

df = df.sample(frac=1)
df
```

```python
x_data = df.iloc[:, :12]

x_data.head()
```

```python
y_data = df.iloc[:, 13]

y_data.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train, y_train)
x_test = sc.transform(x_test)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from numpy import mean, std

rf_clf = ensemble.RandomForestClassifier(n_estimators=100)

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(rf_clf, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print(f'Accuracy: {(mean(n_scores), std(n_scores))}')
```

```python
from sklearn.ensemble import ExtraTreesClassifier


xt_clf = ExtraTreesClassifier()

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

n_scores = cross_val_score(xt_clf, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)

# report performance
print(f'Accuracy: {(mean(n_scores), std(n_scores))}')
```

```python
from xgboost import XGBClassifier

# define the model
XGB_model = XGBClassifier()

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

n_scores = cross_val_score(XGB_model, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)

# report performance
print(f'Accuracy: {(mean(n_scores), std(n_scores))}')
```

```python
from lightgbm import LGBMClassifier

LGM_model = LGBMClassifier()

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(LGM_model, x_train, y_train, scoring='accuracy', cv=cv)
print(f'Accuracy: {(mean(n_scores), std(n_scores))}')
```

```python
from sklearn.model_selection import RandomizedSearchCV


n_estimators = [int(x) for x in np.linspace(start = 1, stop = 20, num = 20)] # number of trees in the random forest
max_features = ['auto', 'sqrt'] # number of features in consideration at every split
max_depth = [int(x) for x in np.linspace(10, 120, num = 12)] # maximum number of levels allowed in each decision tree
min_samples_split = [2, 6, 10] # minimum sample number to split a node
min_samples_leaf = [1, 3, 4] # minimum sample number that can be stored in a leaf node
bootstrap = [True, False] # method used to sample data points

random_grid = {'n_estimators': n_estimators,

'max_features': max_features,

'max_depth': max_depth,

'min_samples_split': min_samples_split,

'min_samples_leaf': min_samples_leaf,

'bootstrap': bootstrap}

rf_random = RandomizedSearchCV(estimator = xt_clf, param_distributions = random_grid, n_iter = 100, scoring="accuracy", cv = 5, verbose=2, random_state=1, n_jobs = -1)
rf_random.fit(x_train, y_train)
random_predictions = rf_random.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix(y_test, random_predictions)
```

```python
classification_report(y_test, random_predictions)
```
